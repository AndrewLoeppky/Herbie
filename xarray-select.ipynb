```python
import xarray as xr

from herbie import Herbie
```

```python
H = Herbie("2024-03-21 00:00", model="hrrr", product="sfc", fxx=0)
ds = H.xarray("TMP:2 m")
```

```python
ds
```

```python
df = ds[["latitude", "longitude"]].to_dataframe().sample(10).reset_index()
df
```

```python
df.y.to_xarray()
```

```python
ds.sel(x=df.x.to_xarray(), y=df.y.to_xarray()).to_pandas()
```

```python
import pandas as pd
import numpy as np

da = xr.DataArray(
    np.random.rand(4, 3),
    [
        ("time", pd.date_range("2000-01-01", periods=4)),
        ("space", ["IA", "IL", "IN"]),
    ],
)


times = xr.DataArray(
    pd.to_datetime(["2000-01-03", "2000-01-02", "2000-01-01"]), dims="new_time"
)
da
```

```python
times
```

```python
da.sel(space=xr.DataArray(["IA", "IL", "IN"], dims=["new_time"]), time=times)
```
