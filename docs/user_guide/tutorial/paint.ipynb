Herbie comes with it's own paint, or custom colormaps for various types of maps. I don't promise they are all "color blind friendly", but they look nice to me.

This includes:

- National Weather Service custom color curves.
- Brian's custom terrain cmaps.


```python
from herbie import Herbie
from herbie.toolbox import EasyMap, pc
import herbie.paint as paint

import matplotlib.pyplot as plt
```

```python
plt.get_cmap("nws_tmp")
```

```python
H = Herbie("2024-01-01")
ds = H.xarray("TMP:2 m")

ax = EasyMap("50m", crs=ds.herbie.crs).STATES().ax
ax.pcolormesh(ds.longitude, ds.latitude, ds.t2m, cmap="nws_tmp", transform=pc)
```

```python
t = H.terrain()
ax = EasyMap(crs=t.herbie.crs, add_coastlines=False).ax
ax.pcolormesh(t.longitude, t.latitude, t.orog, cmap="land_green", transform=pc)
```

```python
ax = EasyMap(crs=t.herbie.crs, add_coastlines=False).ax
ax.pcolormesh(t.longitude, t.latitude, t.orog, cmap="land_tan", transform=pc)
```
